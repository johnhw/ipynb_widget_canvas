## Multi-Frame Image Data

```python
import base64

def encode_src_data(image):
    """
    Compress and encode input image data.
    Format as <img> element src string with embedded data.
    This function uses a PNG compression function from the included module image.py.
    """
    # Compress with PNG.
    data_comp, fmt = canvas.image.png_compress(image)

    # Encode with base64.
    data_b64 = base64.b64encode(data_comp)

    # Build an Image Element src string with embedded data.
    src = 'data:image/{:s};base64,{:s}'.format(fmt, data_b64)
    
    return src
```

```python
# Three images of the little Mini Cooper I keep on my desk.
fnames = ['images/IMG_20140417_065746~2b.jpg',
          'images/IMG_20140417_065746~5b.jpg',
          'images/IMG_20140417_065746~3b.jpg']

# Load image files using my included image processing helper module.
images = [canvas.image.read(f) for f in fnames]

srcs = [encode_src_data(img) for img in images]

# All of my test images have the same width and height.
height, width = images[0].shape[:2]
```

## Simple Animation

This image display tool was never intended to be efficient for animating through a sequence of image frames, but it's still fun to try it and see how it goes. 

```python
wid = canvas.CanvasWidget(srcs[1])
wid
```

```python
num_iter = 100

time_0 = time.time()
for k in range(num_iter):
    # Get a new image, put it in the widget.
    wid.src = srcs[k % num_images]

    
time_1 = time.time()
fps = num_iter / (time_1 - time_0)

print('Frames per second: {:.2f}'.format(fps))
```
